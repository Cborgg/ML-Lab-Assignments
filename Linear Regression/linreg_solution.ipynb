```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

### Learning the model

We had collected the data from a lot of people at NITK sometime back. 
This data is available in htwt.txt. Our task is to learn a linear regression model 
to predict a person's weight from the height. 

```python
x = np.loadtxt('htwt.txt')
plt.plot(x[:, 0], x[:, 1], 'bo')
plt.xlabel('Height --> ')
plt.ylabel('Weight -->')
 
y = x[:, 1]
H = np.column_stack((np.ones(len(x)), x[:, 0]))

a = np.linalg.inv(H.T @ H) @ H.T @ y

   
y_h = H @ a
# Use the closed form formula 
# to find the vector a = [m c]^T 

# Find the predicted values using the 
# equation y_h =  H a   

# plot the predicted values [will be a line ]
plt.plot(x[:,0],y_h,'r-')
    
# Find the Mean Square Error in Train Data




mse = np.mean((y_h - y) ** 2)
print(mse)
plt.plot()


```
